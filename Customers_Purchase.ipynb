```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Mall_Customers.csv')

# One-hot encode categorical variables
data = pd.get_dummies(data)

# Drop customer_id for clustering purposes


# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Elbow Method to determine the number of clusters
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Apply K-means clustering
optimal_clusters = 2
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original data
data['cluster'] = clusters

# Print cluster centroids
print(kmeans.cluster_centers_)

# Analyze the clusters
cluster_analysis = data.groupby('cluster').mean()
print(cluster_analysis)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Mall_Customers.csv')

# One-hot encode categorical variables
data = pd.get_dummies(data)

# Drop customer_id for clustering purposes


# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Elbow Method to determine the number of clusters
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Determine optimal number of clusters (this should be based on the Elbow Method graph)
optimal_clusters = 2  # Adjust based on the Elbow Method result

# Apply K-means clustering
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original data
data['cluster'] = clusters

# Print cluster centroids
print(kmeans.cluster_centers_)

# Analyze the clusters
cluster_analysis = data.groupby('cluster').mean()
print(cluster_analysis)

# Assuming 'Age' and 'Annual Income (k$)' are columns in your DataFrame 'data'
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=data['cluster'], cmap='viridis', label='Data points')

# Plotting centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, data.columns.get_loc('Annual Income (k$)')], centroids[:, data.columns.get_loc('Spending Score (1-100)')],
            s=300, c='red', label='Centroids', marker='X')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100) ')
plt.title('Scatter Plot: Annual Income (k$) vs.Spending Score (1-100) ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('Mall_Customers.csv')

# One-hot encode categorical variables
data = pd.get_dummies(data)

# Drop customer_id for clustering purposes
data = data.drop('CustomerID', axis=1)

# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply PCA to reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Elbow Method to determine the number of clusters
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Determine optimal number of clusters (this should be based on the Elbow Method graph)
optimal_clusters = 2  # Adjust based on the Elbow Method result

# Apply K-means clustering
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(pca_data)

# Add the cluster labels to the PCA data
pca_df = pd.DataFrame(pca_data, columns=['PCA1', 'PCA2'])
pca_df['cluster'] = clusters

# Print cluster centroids in PCA space
print(kmeans.cluster_centers_)

# Scatter plot of PCA components with cluster labels
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['cluster'], cmap='viridis', label='Data points')

# Plotting centroids in PCA space
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids', marker='X')

plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('Scatter Plot: PCA1 vs. PCA2 with Centroids')
plt.legend()
plt.grid(True)
plt.show()
```
